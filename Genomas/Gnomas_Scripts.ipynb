### In this notebook are the diferent scritps I use to get the graphics we need to understand the behavior of the genomes and chromosomas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import altair as alt
```

```python
class GenomaScatter700():

    def decodeTranscript(self,x):
        '''
        This function decode the transcript putting the right name of the gene
        '''
        gen = x.split('_')[0]
        gen_d = {'chr17':'_BRCA1', 'chr16':'_PALB2','chr13':'_BRCA2','chr11':'_ATM','chr2':'_MSH2', }
        return gen + gen_d[gen]


    def clean(self):
        df = pd.read_csv('scater_dataOK_revisado.txt', delimiter='\t')                     # A: Convert the file into a DF
        df.rename(columns={'add2':'Tries','add3':'Genes_N'},inplace=True)                  # A: Change the name of the columns
        df['Genes'] = df['transcript'].apply(lambda x: self.decodeTranscript(x))           # A: Creo columna con valores de transcript

        # Convert the data into categorical data reduce the use of memory from 67.7 kb kb to 41.3kb
        df['sampleID'] = df['sampleID'].astype('category')
        df['add1'] = df['add1'].astype('category')
        df['transcript'] = df['transcript'].astype('category')
        df['Genes'] = df['Genes'].astype('category')

        return df


    def scatter_size(self, df):
        # Conditional variables
        click = alt.selection_multi(encodings=['color'])                                   # U: Can select the genoma and filter the graph
        
        # Chart variables
        x_circle = alt.X('sampleID:O', axis = alt.Axis(title = 'sampleID', grid=True))   # U: The X with de Label Angel
        y_circle = alt.Y('TPM', axis = alt.Axis(title = 'TPM', grid=True))
        
        color_circle = alt.condition(click, alt.Color(shorthand='Genes:N'), alt.value('lightgray'))                         # U: The color variable without the legend     
        size_circle = alt.Size('transcript:N', scale=alt.Scale(range=[150, 200]),legend=None)           # U: The size with the range of scale
        

        
        # Chart base
        circle = alt.Chart(df).mark_circle().encode(x = x_circle, y = y_circle, color = color_circle, size = size_circle     # U: The basic chart
                                          ).properties(height=720                                               # U: The heigth and width
                                          ).add_selection(click)
        line = alt.Chart(df).mark_line().encode(x = x_circle, y = y_circle, color = color_circle, #size = size_circle     # U: The basic chart
                                          ).properties(height=720                                               # U: The heigth and width
                                          )

        
        file = circle + line
        
        
        return file.save('./Plotts/Scatter_Dinamico_700.html') 
    
        
    def graf(self):
        df = self.clean()
        return self.scatter_size(df)

g = GenomaScatter700()
g.graf()
```

```python
def graf(df):
    sns.set(style='darkgrid')                                # A: Seteo el color de fondo
    fig = plt.figure(figsize=(1440/96, 600/96), dpi=96)      # A: Doy tamaño a la figura

    sns.scatterplot(x='sampleID',                            # A: Creo el plot
                    y= 'value', 
                    hue= 'Muestras',
                    data=df)
    
    plt.xlabel('Muestras')                                   # A: Cambio el nombre de la etiqueta
    plt.xticks(rotation=45)                                  # A: Roto las etiquetas para que sean legibles
    plt.xlim(-1,96)                                          # A: Limito el eje x para que tome todo el gráfico
    
    plt.ylabel('Valores')                                    # A: Cambio el nombre de la etiqueta
    
    plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left', ncol=5,    # A: Coloco la leyenda arriba
           mode="expand", borderaxespad=0., ) 
     
    plt.savefig('ejemplo_1', dpi=96)                         # A: Salvo el plot
    
#graf(df)
```

```python
def scatter_line(df):

    # Conditional Variables
    brush = alt.selection_interval(encodings=['x'])  # U: Active on the top panel 
    click = alt.selection_multi(encodings=['color']) # U: Active on the bottom panel 
    
    # Chart Variables
    x_scatter = alt.X('sampleID', axis = alt.Axis(title = 'Muestras', grid = True, labelAngle=-45))      # U: X axis with the data, title, grid and the agle of label
    y_scatter = alt.Y('value', axis = alt.Axis(title = 'Valores', grid = True))                          # U: Y axis with the data, title and grid.
    color_scatter = alt.condition(brush, alt.Color(shorthand='Muestras:N'), alt.value('lightgray'))      # U: Color, with condition, the data, and the value of the not selected data
    
    x_bar = alt.X('value', axis = alt.Axis(title = 'Valores', grid = True))
    y_bar = alt.Y('transcript', axis = alt.Axis(title = '¿Genomas?', grid = False))   
    color_bar = alt.condition(click, alt.Color(shorthand='transcript:N'), alt.value('lightgray'))
    
    # Charts
    scatter = alt.Chart(df).mark_point().encode(x = x_scatter, y = y_scatter,  color = color_scatter     # U: The basic data of the chart
                                       ).properties(width=1280                                           # U: The width, and can go other propertis
                                       ).add_selection(brush                                             # U: The selection for this chart
                                       ).transform_filter(click)                                         # U: The way this chart can be transformed

    bar = alt.Chart(df).mark_bar().encode(x = x_bar, y = y_bar,  color = color_bar 
                                     ).properties(width=600
                                     ).add_selection(click
                                     ).transform_filter(brush)
    
    chart = alt.vconcat(scatter,bar)                                                                     # U: The position of the charts, from up to bottom
    chart.save('Scatter_Bar_Dinamico.html')                                                              # A: Saving the chart
    
    return chart

#scatter_line(df)
```

## Funcion para graficar los Genomas
##### Permite:
- Hacer un gráfico plano con Seaborn donde se muestran los distintos experimentos y sus resultados. 
- Hacer un gráfico interactivo donde se muestran los experimentos, sus resultados y se puede filtrar por genoma. 

```python
class Genoma():
    
    # ================= Función de seleccion de gráfico =================
    def type(self):
        
        for i in range(3):
            
            g_type = float(input('Seleccione el tipo de gráfico que desea realizar: \n[1] Grafico plano con Seaborn \n[2] Grafico interactivo con Altair\n[3] Ambos\n'))

            if g_type == 1:
                print('Eligió "Grafico plano con Seaborn"\nConsulte la carpeta que el gráfico debe encontrarse allí')
                return g_type
            elif g_type == 2:
                print('Eligió "Grafico interactivo con Altair"\nConsulte la carpeta que el gráfico debe encontrarse allí')
                return g_type
            elif g_type == 3:
                print('Eligió "Ambos"\nConsulte la carpeta que los gráfico deben encontrarse allí')
                return g_type
            else:
                print('Elija una opción válida\n')
                
        print('Lo siento, ninguna de las opciones fue válida')
    
    
    # ================ Función de limpieza de Data Frame ================
    def clean(self):
        df = pd.read_csv('scater_dataOK_revisado.txt', delimiter='\t')            # A: Tomo el archivo 
        df.columns = ["sampleID", "add1", "transcript", "add2", "TPM", "add3"]    # A: Doy nombre a las columnas
        df['addT'] = df['transcript'].apply(lambda x: x[-1])                      # A: Creo columna con valores de transcript
        df['TPM'] = df['TPM'].div(10000)                                          # A: Reduzco el valor para que entre en el gráfico  

        df = df.melt(id_vars=['sampleID','transcript'], var_name='Muestras',      # A: Armo el DF para poder graficar
                     value_vars=['add2','add3','addT','TPM'])


        return df

    
    # ==================== Función de Grafico Plano =====================
    def scatter_seaborn(self, df):
        sns.set(style='darkgrid')                                # A: Seteo el color de fondo
        fig = plt.figure(figsize=(1440/96, 600/96), dpi=96)      # A: Doy tamaño a la figura

        sns.scatterplot(x='sampleID',                            # A: Creo el plot
                        y= 'value', 
                        hue= 'Muestras',
                        data=df)

        plt.xlabel('Muestras')                                   # A: Cambio el nombre de la etiqueta
        plt.xticks(rotation=45)                                  # A: Roto las etiquetas para que sean legibles
        plt.xlim(-1,96)                                          # A: Limito el eje x para que tome todo el gráfico

        plt.ylabel('Valores')                                    # A: Cambio el nombre de la etiqueta

        plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left', ncol=5,    # A: Coloco la leyenda arriba
               mode="expand", borderaxespad=0., ) 

        plt.savefig('Scatter_Bar_Plano', dpi=96)                 # A: Salvo el plot

        return

    
    # ================= Función de Grafico Interactivo ==================
    def scatter_altair(self, df):

        # Conditional Variables
        brush = alt.selection_interval(encodings=['x'])  # U: Active on the top panel 
        click = alt.selection_multi(encodings=['color']) # U: Active on the bottom panel 

        # Chart Variables
        x_scatter = alt.X('sampleID', axis = alt.Axis(title = 'Muestras', grid = True, labelAngle=-45))      # U: X axis with the data, title, grid and the agle of label
        y_scatter = alt.Y('value', axis = alt.Axis(title = 'Valores', grid = True))                          # U: Y axis with the data, title and grid.
        color_scatter = alt.condition(brush, alt.Color(shorthand='Muestras:N'), alt.value('lightgray'))      # U: Color, with condition, the data, and the value of the not selected data

        x_bar = alt.X('value', axis = alt.Axis(title = 'Valores', grid = True))
        y_bar = alt.Y('transcript', axis = alt.Axis(title = 'Genomas', grid = False))   
        color_bar = alt.condition(click, alt.Color(shorthand='transcript:N'), alt.value('lightgray'))

        # Charts
        scatter = alt.Chart(df).mark_point().encode(x = x_scatter, y = y_scatter,  color = color_scatter     # U: The basic data of the chart
                                           ).properties(width=1280                                           # U: The width, and can go other propertis
                                           ).add_selection(brush                                             # U: The selection for this chart
                                           ).transform_filter(click)                                         # U: The way this chart can be transformed

        bar = alt.Chart(df).mark_bar().encode(x = x_bar, y = y_bar,  color = color_bar 
                                         ).properties(width=600
                                         ).add_selection(click
                                         ).transform_filter(brush)

        chart = alt.vconcat(scatter,bar)                                                                     # U: The position of the charts, from up to bottom
        chart.save('Scatter_Bar_Dinamico.html')                                                              # A: Saving the chart

        return chart

    
    # ================ Función de Selección de Gráfico ==================
    def  graf(self):
        
        df = self.clean()
        sel = self.type()
        
        if sel == 1:
            return self.scatter_seaborn(df)
        elif sel == 2:
            return self.scatter_altair(df)
        elif sel == 3:
            return self.scatter_seaborn(df), self.scatter_altair(df)
        else:
            print('No se como llegaste hasta acá, estás haciendo trampa... ¬¬')
        
```
